```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.plot(df['total_bill'])
```

```python
df.sort_values(by='total_bill',ascending=True,inplace=True)
df.head()
```

```python
print(f"Minimum bill : {df['total_bill'].min()}")
print(f"Maximum bill : {df['total_bill'].max()}")
sns.displot(data=df,x=df['total_bill'],kde=True)
plt.grid()
```

```python
sns.displot(data=df,x='tip')
plt.grid()
```

```python
df.groupby(by='time').mean(numeric_only=True) #People tend to give tip more at dinner as compared to lunch
```
